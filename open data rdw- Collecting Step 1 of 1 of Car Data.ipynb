```python
import pandas as pd


# Read the CSV file into a DataFrame
rdw_data = pd.read_csv('/Users/feysal/Downloads/Open_Data_RDW__Gekentekende_voertuigen_20250110.csv')

# Filter the RDW dataset for passenger cars (Personenauto)
personenauto_df = rdw_data[rdw_data['Voertuigsoort'] == 'Personenauto']

# Display the shape of the filtered dataset
print(f"Filtered dataset for Personenauto: {personenauto_df.shape}")

```

```python
# Count occurrences of Handelsbenaming (car models)
handelsbenaming_counts = personenauto_df['Handelsbenaming'].value_counts()

# Get the top 10 most frequent Handelsbenaming
top_10_models = handelsbenaming_counts.head(10)

# Visualize the top 10 car models
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
top_10_models.plot(kind='bar', color='skyblue')
plt.title("Top 10 Most Popular Car Models (Handelsbenaming)", fontsize=16)
plt.xlabel("Car Model (Handelsbenaming)", fontsize=14)
plt.ylabel("Number of Registrations", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Count occurrences of Merk (brands)
brand_counts = personenauto_df['Merk'].value_counts()

# Get the top 10 most frequent brands
top_10_brands = brand_counts.head(10)

# Visualize the top 10 car brands
plt.figure(figsize=(12, 6))
top_10_brands.plot(kind='bar', color='orange')
plt.title("Top 10 Most Popular Car Brands", fontsize=16)
plt.xlabel("Brand (Merk)", fontsize=14)
plt.ylabel("Number of Registrations", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Convert 'Datum eerste toelating DT' to datetime format
personenauto_df['Datum eerste toelating DT'] = pd.to_datetime(personenauto_df['Datum eerste toelating DT'], errors='coerce')

# Extract year from the registration date
personenauto_df['Registration Year'] = personenauto_df['Datum eerste toelating DT'].dt.year

# Group data by year and count the registrations
registrations_by_year = personenauto_df.groupby('Registration Year').size()

# Plot the trend of registrations over the years
plt.figure(figsize=(12, 6))
registrations_by_year.plot(kind='line', marker='o', color='green')
plt.title("Trend of Vehicle Registrations Over Time", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Registrations", fontsize=14)
plt.grid()
plt.show()
```
