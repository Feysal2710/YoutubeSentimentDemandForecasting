```python
import requests
import pandas as pd
import googleapiclient.discovery


API_KEY = 'Hidden'  

# Define all car models and their video IDs
car_models = {
    "Tesla Model Y": {
        "English": ["PZ8NPeYFPCY", "9ylI0r7L78o"],
        "Dutch": ["E5mfhe-Q6lE", "PugfhRhRNaI", "6472WM0QfmA"]
    },
    "Kia Picanto": {
        "English": ["RvHRyIRHbw"],
        "Dutch": ["WVRdVRZSatI", "cgUImTjPo9A"]
    },
    "Peugeot 208": {
        "English": ["rhn8o1zAzdg", "2ajatwrK1sw"],
        "Dutch": ["1BrvV2C4km4", "W98k1x44WFU", "P7AyEy_RyA4"]
    },
    "Volvo XC40": {
        "English": ["dDnylpt9B8I"],
        "Dutch": ["__QoKqx3MlE", "c44K-eyEGKE"]
    },
    "Kia Niro": {
        "English": ["L1ec7qEhthA"],
        "Dutch": ["F1bHv9LtkhA", "5aV37uUn-tU"]
    },
    "Volkswagen Polo": {
        "English": ["WyawJQPxQ4M"],
        "Dutch": ["uf35KjAETmQ", "mQPBhpEl_vg"]
    },
    "Opel Corsa": {
        "English": ["p2BwYecNWrg"],
        "Dutch": ["rwm90uurYxo", "Ve1ffG4Rsdo", "oDJ8kQSA7EI"]
    },
    "Lynk & Co 01": {
        "English": ["R4YSOsDfQV8"],
        "Dutch": ["I_U3j0lhAqw", "gLu3ZpBmw1w", "ERRUxou7VNk", "gmwQbxtnlOM"]
    },
    "Toyota Yaris Cross": {
        "English": ["wAwtbHGXnhc", "X-cFdTKFBQE"],
        "Dutch": ["XIbS4zd2h-Q", "0ZJrQUA6z90"]
    },
    "Toyota Aygo": {
        "English": ["DWuLJhUcJhY"],
        "Dutch": ["sRK4mBA26wo", "t9nXoYAN1Ag", "I_YYIOIQDFU"]
    },
    "Renault MÃ©gane": {
        "English": ["2GJMyKFAM-I"],
        "Dutch": ["V2vyYY3dC1w", "kgO4iywz9vE"]
    },
    "Tesla Model 3": {
        "English": ["F8KUdvmW2EQ"],
        "Dutch": ["bDJ5ZqN4FYE", "YDdPkq0hXCs"]
    },
    "Peugeot 2008": {
        "English": ["iKN6lpFsB6g"],
        "Dutch": ["_8SWSGV8KwA", "q__1pWFOLEw"]
    },
    "Skoda Enyaq": {
        "English": ["2rmYL3_mwP4"],
        "Dutch": ["YnfhTjQ5DGg", "ZZdH_95q7ys"]
    },
    "Renault Clio": {
        "English": ["v_86Cb2iHmM", "u4sR-EGP4LU"],
        "Dutch": ["hlqRtehdjkY", "w8OsY7ssokg", "NXWaa1tKaeE", "3my4GFo_VtU"]
    },
    "Toyota Yaris": {
        "English": ["FhdWqdFzvGM"],
        "Dutch": ["T1cvTkSrLyM"]
    },
    "Volkswagen Up": {
        "English": ["ByJntRTrbUk"],
        "Dutch": ["4DwEFQ5uebo"]
    },
    "Kia Sportage": {
        "English": ["0Ghjlt1GvSg"],
        "Dutch": ["Y0ekM5XH1Aw", "M7r6EyhLKRo", "X0g0tzEnuhs"]
    },
    "Fiat 500": {
        "English": ["tpOSwBjgj74"],
        "Dutch": ["v0nxStk-lmI", "FhmcZA661aw", "9jvnjrY0b4s"]
    },
    "Volkswagen Golf": {
        "English": ["E2cr8Xkg_KI"],
        "Dutch": ["mY4k8WXRcF4", "Z_2sJhIDe_M"]
    },
    "Mercedes-Benz C-Class": {
        "English": ["02yxl06J4Uw"],
        "Dutch": ["Z_aZ-tutZrY", "HSwodmvPO6c"]
    },
    "Volvo EX30": {
        "English": ["5umdHj0o1B8"],
        "Dutch": ["V1mSiz4ujos", "iG8waSJCpqk"]
    },
    "BMW X1": {
        "English": ["fEbunz6ed0Q"],
        "Dutch": ["U0t2k7kFY-U", "NmHxPgVVEYg", "aEH5Pz6nK78"]
    },
    
    "Toyota Corolla": {
        "English": ["Y0tcDV_50hk"],
        "Dutch": ["EE7LYGSNp7Y", "tatw_cNih6s"]
    },
    "BMW X3": {
        "English": ["gfXVwawi_Ik"],
        "Dutch": ["DHwRNd5K_EI"]
    },
    "Mercedes-Benz A-Class": {
        "English": ["wc4QV7yTAPc"],
        "Dutch": ["EMQYN0dRFZY", "2sv4s_keKe8"]
    },
    "Honda CR-V": {
        "English": ["Q6NJDHcstXw"],
        "Dutch": ["LXDkYqzmfdk", "uZqDhmidfdw"]
    },
    "Hyundai Kona": {
        "English": ["VhHwZub5-bY"],
        "Dutch": ["qVuUlH5O6sE", "h3_ulyLBzT4"]
    },
    "Skoda Octavia": {
        "English": ["SErOkO4UIXE"],
        "Dutch": ["EuXKw9jJWhY", "QEUuKw53M9A"]
    },
    "BMW X5": {
        "English": ["5nQCX_ygeA0"],
        "Dutch": ["EuzMZWHsPKY", "DZVLtGal4Eo"]
    },
    "Skoda Superb": {
        "English": ["pGfGeT23lc8"],
        "Dutch": ["Sp36thgFii8", "72DRCkTlG0M", "HW5RrbKMt8o"]
    },
    "Skoda Fabia": {
        "English": ["EKszuyDoQZo"],
        "Dutch": ["Fvyg7ATxI-k", "4cB8zCTDtQQ"]
    },
    "Audi Q4 E-tron": {
        "English": ["5puyCvtaJEQ"],
        "Dutch": ["tu_1eZvYof0", "8uusZCYKrKI"]
    },
    "Hyundai i10": {
        "English": ["PFZhMpFytFU"],
        "Dutch": ["gLA8HgIPcIw", "SEyQT7WdSVw"]
    },
    "Volvo XC60": {
        "English": ["o4MYseY9rOk"],
        "Dutch": ["3uVIVed_ojM", "IumhcNDfMvw"]
    },
    "Ford Kuga": {
        "English": ["uCUNgoLxXQo"],
        "Dutch": ["1mMcDliZgjY", "gv87AaB_GG0"]
    },
    "Volkswagen Tiguan": {
        "English": ["cTLOXDTdUPw"],
        "Dutch": ["sJqCf8ceElk", "elbQ-rdKTgQ"]
    },
    "MINI Cooper": {
        "English": ["Bg9R2iVSlMo"],
        "Dutch": ["DCepHrnxy8s", "gz_NEq2MkHQ"]
    },
    "Renault Scenic": {
        "English": ["MfXP8kj5s_M"],
        "Dutch": ["ySDtEOKnNo0", "GKaF2BK5j7M"]
    },
    "Mercedes-Benz E-Class": {
        "English": ["vaDuKbiL9NU"],
        "Dutch": ["HAZIa8Zdpko", "_AFMdm7gAdo"]
    }
}


def fetch_all_comments(video_id):
    """
    Fetches all comments for a given video ID using the YouTube Data API.
    """
    comments = []
    url = f"https://www.googleapis.com/youtube/v3/commentThreads?part=snippet&videoId={video_id}&maxResults=100&key={API_KEY}"
    
    while url:
        response = requests.get(url).json()
        if "items" in response:
            for item in response['items']:
                comment = item['snippet']['topLevelComment']['snippet']
                comments.append([
                    comment.get('authorDisplayName', 'Unknown'),
                    comment.get('publishedAt', 'N/A'),
                    comment.get('updatedAt', 'N/A'),
                    comment.get('likeCount', 0),
                    comment.get('textDisplay', '')
                ])
        # Check if there is a next page
        url = f"https://www.googleapis.com/youtube/v3/commentThreads?pageToken={response.get('nextPageToken')}&part=snippet&videoId={video_id}&maxResults=100&key={API_KEY}" if 'nextPageToken' in response else None
    return comments

# Fetch comments for all car models with pagination
data = []

for car_model, languages in car_models.items():
    for language, video_ids in languages.items():
        for video_id in video_ids:
            comments = fetch_all_comments(video_id)
            for comment in comments:
                data.append({
                    "Car Model": car_model,
                    "Language": language,
                    "Video ID": video_id,
                    "Author": comment[0],
                    "Published At": comment[1],
                    "Updated At": comment[2],
                    "Like Count": comment[3],
                    "Comment Text": comment[4]
                })

# Convert to a DataFrame
datacomments = pd.DataFrame(data)

# Save the full dataset
datacomments.to_csv("car_model_comments_full_new.csv", index=False)

# Separate datasets for Dutch and English comments
dutch_comments_new = datacomments[datacomments['Language'] == 'Dutch']
english_comments_new = datacomments[datacomments['Language'] == 'English']

# Save separate datasets
dutch_comments_new.to_csv("dutch_comments_full_new.csv", index=False)
english_comments_new.to_csv("english_comments_full_new.csv", index=False)
```
